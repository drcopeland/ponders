```python
def find_lengths(end):
    for i in range(1, end+1):
        found_prime = False
        length = 1
        a = i
        n = 1
        while found_prime == False:
            if is_prime(a):
                
                length = n
                found_prime = True
                break
            a = a + n
            n = n + 1
        print(i, length)
    return
                
```

```python
find_lengths(300)
```
